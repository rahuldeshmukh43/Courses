```python
#Defining functions for various problems
import numpy as np

# Code for infinity norm of a matrix
def inf(a):
    x=(abs(a).sum(axis=1).max())
    return x
# Code for Gauss Elimination with Partial Pivoting
def Gaussp(A):
    n = len(A)
    for i in range(0, n):
        maxtemp = abs(A[i][i])
        pivotrow = i
        
        for k in range(i+1,n):
            if (abs(A[k][i])>maxtemp):
                maxtemp=abs(A[k][i])
                pivotrow=k
                
        for k in range(i,n+1):
            temp=A[i][k]
            A[i][k]=A[pivotrow][k]
            A[pivotrow][k]=temp
            
        for k in range(i+1,n):
            c=-(A[k][i])/(A[i][i])
           # if (c<1e-10):
            #    c=0
            for j in range(i, n+1):
                if i == j:
                    A[k][j] = 0
                else:
                    A[k][j] += c * A[i][j]

        
    x = [0 for i in range(n)]
    for i in range(n-1, -1, -1):
        x[i] = A[i][n]/A[i][i]
        for k in range(i-1, -1, -1):
            A[k][n] -= A[k][i] * x[i]
    return x

#Code for Gauss Elimination with scaled pivoting
def Gscale(v):
    n=len(v)
    row=np.empty([n,1])
    s=(np.abs(v)[:,:-1]).max(axis=1)
    
    for i in range(0,n):
        for k in range(i,n):
            row[k]=v[k][i]/s[k]
            temp=abs(row[i])
            srow=i
        for k in range(i,n):
            if (abs(row[k])>temp):
                temp=abs(row[k])
                srow=k
        for k in range(i,n+1):
            temp=v[i][k]
            v[i][k]=v[srow][k]
            v[srow][k]=temp
        for k in range(i+1,n):
            c=-v[k][i]/v[i][i]
            #if(c<(1e-10)):
             #   c=0
            for j in range(i,n+1):
                v[k][j]=v[k][j]+(c*v[i][j])
    y = [0 for i in range(n)]
    for i in range(n-1, -1, -1):
        y[i] = v[i][n]/v[i][i]
        for k in range(i-1, -1, -1):
            v[k][n] -= v[k][i] * y[i]
    return y
```

## Problem 1

A system of equations Ax=b is given as

$\left[\begin{array}{cc}2.01 & 1.99 \\1.99 & 2.01  \\ \end{array} \right]$$x$=$\left[\begin{array}{c}4\\4\\\end{array} \right]$

The solution to this system of equations is
$x=\left[ \begin{array}{c}1\\1\\\end{array} \right]$  ,
and one approximate solution is
$\widetilde{x}=\left[ \begin{array}{c}2\\0\\ \end{array} \right]$ .<br>
1) Compute the error $e=x-\widetilde{x}$

2) Compute the residual $r=A\widetilde{x}-b$.

3) Using the ${l}_\infty$ norm, compute the relative error $\frac{{||e||}_\infty}{{||x||}_\infty}$.

4)	Using the ${l}_\infty$, compute the condition number ${k}_\infty$.

5)	Using the ${l}_\infty$, compute the relative residual $\frac{{||r||}_\infty}{{||b||}_\infty}$.

6)	Compute the product of the condition number and the relative residual.

7)	Compare the relative error to the product of the condition number and the relative residual.

### Solution

```python
x1=np.matrix([1,1])
x2=np.matrix([2,0])
A1=np.matrix([[2.01,1.99],[1.99,2.01]])
B1=np.matrix([4,4])

#1)
e=x1-x2
print("1) The error e = ",e)

#2)
r=(A1*x2.T)-B1
print("\n2) The residual r = \n",r)

#3)
r2=inf(e)/inf(x1)
print("\n3) The e/x ratio using the l infinity norm is ",r2)

#4)
k=inf(A1)*inf(A1.I)
print("\n4) The condition number k =",k)

#5)
r3=inf(r)/inf(B1)
print("\n5) The relative residual r/b using the l infinity norm is ",r3)

#6)
print("\n6) The product of the condition number and the relative residual is ",k*r3)


```

7) The relative error is seen to be equal than the product of the condition number and the relative residual Thus it satisfies the condition:
$\dfrac{{||e||}}{{||x||}} \leqslant ||{A}||*||{{A}^{-1}}||*\dfrac{{||r||}}{{||b||}}$ 

## Problem 2
Let $\qquad$A=$\left[ \begin{array}{cccc}5.1 & 8.7 \\2.4 & 4.1  \\ \end{array} \right]\qquad$ B=$\left[ \begin{array}{c}9.48\\4.48\\\end{array} \right]$<br><br>
1)	Using the ${l}_\infty$ norm, compute the condition number ${k}_\infty(A)$.<br>

2)	Solve the system of equations $Ax=b$ for $x$.<br>

3)	Perturb the coefficient matrix A and the right-side vector b by
$δA=\left[ \begin{array}{rr}-0.001 & 0 \\0.001 & 0  \\ \end{array} \right]$   and   $δb=\left[ \begin{array}{cc}0.05\\-0.05\\\end{array} \right]$
and solve the resulting perturbed system of equations $(A+δA)\widetilde{x}=(b+δb)$ for the approximate solution $\widetilde{x}$.<br>

4)	Using the ${l}_\infty$ norm, compute the actual value of the relative change in the solution,$\dfrac{{||δx||}_\infty}{||x||_\infty}$ for the perturbation in part (3).<br>

5)	Using the ${l}_\infty$ norm, compute the theoretical upper bound of the relative change in the solution,$\dfrac{{||δx||}_\infty}{||x||_\infty}$ for the perturbation in part (3).<br>

6)	For the perturbation in part (3), compare the actual value of $\dfrac{{||δx||}_\infty}{||x||_\infty}$  to its theoretical upper bound.<br>

7)  Perturb the coefficient matrix A and the right-side vector b by
$δA=\left[ \begin{array}{rrrr}0.001 & -0.001 \\-0.001 & 0.001  \\ \end{array} \right]$   and   $δb=\left[ \begin{array}{cccc}-0.1\\0.1\\\end{array} \right]$
and solve the resulting perturbed system of equations $(A+δA)\widetilde{x}=(b+δb)$ for the approximate solution $\widetilde{x}$.<br>

8) Using the ${l}_\infty$ norm, compute the actual value of the relative change in the solution, $\dfrac{{||δx||}_\infty}{||x||_\infty}$ for the perturbation in part (7).<br>

9)	Using the ${l}_\infty$ norm, compute the theoretical upper bound of the relative change in the solution,$\dfrac{{||δx||}_\infty}{||x||_\infty}$ for the perturbation in part (7).<br>

10)	For the perturbation in part (7), compare the actual value of $\dfrac{{||δx||}_\infty}{||x||_\infty}$  to its theoretical upper bound.<br>

### Solution

```python
A3=np.matrix([[5.1,8.7],[2.4,4.1]])
B3=np.matrix([[9.48],[4.48]])
dA1=np.matrix([[-0.001,0],[0.001,0]])
dB1=np.matrix([[0.05],[-0.05]])
dA2=np.matrix([[0.001,-0.001],[-0.001,0.001]])
dB2=np.matrix([[-0.1],[0.1]])

#1)
m1=inf(A3)*inf(A3.I)
print("\n1) The condition number k is ",m1)

#2)
ans=(A3.I)*(B3)
print("\n2) Solution for the Ax=b is given by x=\n",ans)

#3)
ans1=((A3+dA1).I)*(B3+dB1)
print("\n3) Solution for the first perturbed system is given by x1=\n",ans1)

#4)
ans2=inf(ans1-ans)/inf(ans)
print("\n4) The relative change in solution using the infinity norm is ",ans2)

#5)
up= (m1*((inf(dB1)/inf(B3))+(inf(dA1)/inf(A3))))/(1-(m1*(inf(dA1)/inf(A3))))
print("\n5) The theoretical upper bound for the first perturbed system is ",up)
```

6) 	The relative change in the solution is seen to be less than its theoretical upper bound:<br><br>
$(\dfrac{{||δx||}_\infty}{||x||_\infty}=9.59)< 54.91$<br><br>

```python
#7)
ans3=((A3+dA2).I)*(B3+dB2)
print("\n7) Solution for the second perturbed system is given by x2=\n",ans3)

#8)
ans4=inf(ans3-ans)/inf(ans)
print("\n8) The relative change in solution using the infinity norm is ",ans4)

#9)
up1= (m1*((inf(dB2)/inf(B3))+(inf(dA2)/inf(A3))))/(1-(m1*(inf(dA2)/inf(A3))))
print("\n9) The theoretical upper bound for the second perturbed system is ",up1)
```

10) The relative change in the solution is seen to be less than its theoretical upper bound:<br><br>
$(\dfrac{{||δx||}_\infty}{||x||_\infty}=6.58)< 429.3$<br><br>
